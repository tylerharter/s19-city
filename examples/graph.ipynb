```python
g = Digraph()

g.node('Rock')
g.node('Paper')
g.node('Scissors')

g.edge('Rock', 'Scissors', color="red")
e = g.edge('Scissors', 'Paper', penwidth="5")
g.edge('Paper', 'Rock', label="example-label")

g
```
